```python
#Import the main principals 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#We load dataset to use
adult = pd.read_csv('/content/adult.csv', sep=",")
```

##Data Exploratory

```python
#Visualization of the DataFrame
adult.head(20)
```

```python
headers = ['Age', 'Workclass', 'fnlwgt', 'Education', 'Education-num', 'Marital-status', 'Occupation', 'Relationship', 'Race', 'Sex', 'capital-gain', 'capital-loss', 'Hours-per-week', 'Native-country', 'Salary']

adult.columns = headers
```

```python
adult.head(10)
```

```python
# We eliminate the columns that do not interest us

columns_to_drop = ['fnlwgt', 'capital-gain', 'capital-loss']
adult.drop(columns_to_drop, axis="columns", inplace=True)
```

```python
adult.head(15)
```

```python
# Let's analize the shape of the object
print(adult.shape)
```

```python
# Let's visualize the types of data
adult.dtypes
```

```python
adult['Salary'].value_counts()
```

```python
# Transform data with label encoding
from sklearn.preprocessing import LabelEncoder

lab_en = LabelEncoder()

adult['Workclass'] = lab_en.fit_transform(adult['Workclass'])
adult['Education'] = lab_en.fit_transform(adult['Education'])
adult['Marital-status'] = lab_en.fit_transform(adult['Marital-status'])
adult['Occupation'] = lab_en.fit_transform(adult['Occupation'])
adult['Relationship'] = lab_en.fit_transform(adult['Relationship'])
adult['Race'] = lab_en.fit_transform(adult['Race'])
adult['Native-country'] = lab_en.fit_transform(adult['Native-country'])
```

```python
# One-hot encoding to Sex object
adult_onehot = pd.get_dummies(adult['Sex'])
adult = pd.concat([adult, adult_onehot], axis=1)
adult.drop('Sex', axis=1, inplace=True)
```

```python
adult.head()
```

```python
# One-hot encoding to Salary object
adult_onehot = pd.get_dummies(adult['Salary'])
adult = pd.concat([adult, adult_onehot], axis=1)
adult.drop('Salary', axis=1, inplace=True)
```

```python
coldrop = [' <=50K', ' Female']
adult.drop(coldrop, axis="columns", inplace=True)
```

```python
adult.rename(columns={' >50K': 'Salary'}, inplace=True)
adult.rename(columns={' Male': 'Genre'}, inplace=True)
```

```python
adult.dtypes
```

## Training of Decision Trees with scikit-learn

```python
# Let's separate in X and y
X = adult.drop('Salary', axis=1)
y = adult.Salary
```

```python
# Import the necessary librarie to the creation of the model
from sklearn.model_selection import train_test_split

# 30% to test and 70% to train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=00000)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier

# Creation of the model
tree = DecisionTreeClassifier(max_depth=2, random_state=00000)
```

```python
# Training
tree.fit(X_train, y_train)
```

## Model Test

```python
# Calculation of the predictions in Train and Test
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
# Calculation of metrics
from sklearn.metrics import accuracy_score

# Calculation of accuracy in Train
train_accuracy = accuracy_score(y_train, y_train_pred)

# Calculation accuracy in Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print("The accuracy in train is:", train_accuracy)
print("The accuracy in test is:", test_accuracy)
```

```python
#Let's verify the feature importances
import seaborn as sns

importances = tree.feature_importances_
columns = X.columns
sns.barplot(x=columns, y=importances, saturation=2.0, edgecolor='black', linewidth=2)
plt.xticks(rotation=45, horizontalalignment="right", fontweight="light",  fontsize="x-large")
plt.title("Importance of each Feature")
plt.show()
```

## Training of Classification Model with Random Forest

```python
# Import random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10, random_state=0)
rf.fit(X_train, y_train)
```

```python
#Calculation of the predictions in Train and Test
y_train_pred_rf = rf.predict(X_train)
y_test_pred_rf = rf.predict(X_test)
```

```python
#Calculation of metrics 
from sklearn.metrics import accuracy_score

#Calculation of accuracy in Train
train_accuracy_rf = accuracy_score(y_train, y_train_pred_rf)

#Calculation of accuracy in Test
test_accuracy_rf = accuracy_score(y_test, y_test_pred_rf)

print("The accuracy in train is:", train_accuracy_rf)
print("The accuracy in test is:", test_accuracy_rf)

#Important: We could reduce the number of estimators to para decrease the model overfitting
```

```python
# Visualization of the important feature 
features_score = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
features_score
```

```python
import matplotlib.pyplot as plt

sns.barplot(x=features_score, y=features_score.index)
plt.xlabel("Features Importances Score")
plt.ylabel("Features")
plt.title("Visualizing the Features Importances")
plt.show()
```

```python
# Confusion Matrix of Random Forest
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_test_pred_rf)

print("Confusion Matrix\n\n", cm)
```

```python
#RF
from sklearn.metrics import classification_report

print(classification_report(y_test, y_test_pred_rf))
```
